```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
#%matplotlib inline
```

```python
dataset = pd.read_csv('D:/EverythingD/01SRH-BDBA Acads/Blk7-DataStoryTelling/Data4Analysis/regressionTesting/testDummyData1.csv')
```

```python
dataset.plot(x='var1', y='Y', style='o')  
plt.title('Y vs var1')  
plt.xlabel('var1')  
plt.ylabel('Y')  
plt.show()
```

```python
dataset.plot(x='var2', y='Y', style='o')  
plt.title('Y vs var2')  
plt.xlabel('var2')  
plt.ylabel('Y')  
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
seabornInstance.distplot(dataset['var1'])
```

```python
x1 = dataset['var1'].values.reshape(-1,1)
x2 = dataset['var2'].values.reshape(-1,1)
y = dataset['Y'].values.reshape(-1,1)
```

```python
x1train, x1test, x2train, x2test, ytrain, ytest = train_test_split(x1, x2, y, test_size=0.2, random_state=0)
```

```python
### 
### simple linear reg
###
```

```python
regressor = LinearRegression()  
regressor.fit(x1train, ytrain) #training the algorithm
```

```python
#To retrieve the intercept:
print(regressor.intercept_)
#For retrieving the slope:
print(regressor.coef_)
```

```python
### 
### MULTIPLE linear reg
###
```

```python
xml = dataset[['var1', 'var2']].values
yml = dataset['Y'].values
```

```python
xmltrain, xmltest, ymltrain, ymltest = train_test_split(xml, yml, test_size=0.2, random_state=0)
```

```python
regressorml = LinearRegression()  
regressorml.fit(xmltrain, ymltrain)
```

```python
coeff_df = pd.DataFrame(regressorml.coef_, xml.columns, columns=['Coefficient'])  
coeff_df
```
